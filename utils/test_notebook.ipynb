```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
valeurs = [10, 20, 15]
pourcentages = [25, 50, 25]

fig, ax = plt.subplots()
bars = ax.barh(categories, valeurs)

# Annotations
for i, bar in enumerate(bars):
    width = bar.get_width()
    ax.text(width + 0.5, bar.get_y() + bar.get_height() / 2,
            f"{valeurs[i]} ({pourcentages[i]}%)",
            va='center', ha='left', color='black', fontsize=12)

plt.show()
```
